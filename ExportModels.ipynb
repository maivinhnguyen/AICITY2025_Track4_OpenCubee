```python
TENSORRT_TAR_FILE = "/content/vis_fish_finetune_dfine/TensorRT-10.10.0.31.Linux.x86_64-gnu.cuda-12.9.tar.gz"

print(f"Extracting {TENSORRT_TAR_FILE}...")
!tar -xzf {TENSORRT_TAR_FILE} -C /content/
print("Extraction complete.")

# --- IMPORTANT: Identify the extracted directory name ---
# List the contents of /content to find the exact name
print("\n--- Contents of /content/ after extraction: ---")
!ls /content/
```

```python
import os

# --- !!! UPDATE THIS PATH BASED ON STEP 3's OUTPUT !!! ---
# Example: If ls showed 'TensorRT-8.6.1.6', use that.
TENSORRT_VERSION_DIR = "TensorRT-10.10.0.31" # <-- CHANGE THIS
TENSORRT_DIR = f"/content/{TENSORRT_VERSION_DIR}"

# Check if the directory exists to avoid errors
if not os.path.isdir(TENSORRT_DIR):
    print(f"ERROR: Directory {TENSORRT_DIR} not found!")
    print("Please check the output of step 3 and update TENSORRT_VERSION_DIR.")
else:
    print(f"Using TensorRT directory: {TENSORRT_DIR}")

    # Add TensorRT bin directory to PATH
    new_path = f"{TENSORRT_DIR}/bin:" + os.environ.get('PATH', '')
    os.environ['PATH'] = new_path
    print(f"Updated PATH: {os.environ['PATH']}") # For verification

    # Add TensorRT lib directory to LD_LIBRARY_PATH
    new_ld_path = f"{TENSORRT_DIR}/lib:" + os.environ.get('LD_LIBRARY_PATH', '')
    os.environ['LD_LIBRARY_PATH'] = new_ld_path
    print(f"Updated LD_LIBRARY_PATH: {os.environ['LD_LIBRARY_PATH']}") # For verification

    print("\nEnvironment variables set.")
```
